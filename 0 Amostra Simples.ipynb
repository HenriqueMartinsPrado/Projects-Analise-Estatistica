#### Criando a variável amostra para gerar os números aleatórios

```python
# Para garantir que os valores gerados sejam sempre os mesmos, deve-se utilizar o seed
np.random.seed(2345)
```

```python
amostra = np.random.choice(a = [0, 1], size = 150, replace = True, p = [0.5, 0.5])
amostra
```

#### Comandos adicionais

```python
# Quantidade de amostras geradas
len(amostra)
```

```python
# Quantidade de amostras geradas com o valor igual a 1
len(amostra[amostra == 1])
```

```python
# Quantidade de amostras geradas com o valor igual a 0
len(amostra[amostra == 0])
```
